```python
df1.head()
```

```python
# 1) How many camplaings in the dataset
print("The total no. of campaigns are: ", len(df1['campaign'].unique()))
```

```python
# 2) How many users do we have with houseing and personal loan
housingloan = df1[df1['housing'] == 'yes']
personalloan = housingloan[housingloan['loan'] == 'yes']['loan'].count()

print("The people who have both housing and personal loan are ", personalloan)
```

```python
# 3) No. of persons with age 60+
print("The people who have 60+ age are ", df1[df1['age'] > 60]['age'].count())
```

```python
# 4) Which month most no. of customer targetted
monthcampaign = df1[['month', 'campaign']]
montharray = monthcampaign['month'].unique()

max = 0
maxMonth = ''
result = []
for i in montharray:
    local_max = monthcampaign[monthcampaign['month'] == i]['campaign'].sum()
    result.append([i, local_max])
    if local_max > max:
        max = local_max
        maxMonth = i
print(maxMonth.upper(), "was the month most customers were tagettted in and the peak target count is", max)
resultDF = pd.DataFrame(result)
print(resultDF[1].max())
```

```python
# 5) WHICH MODE OF CALL GIVE YOU A GOOD RESULT
modeExtraction = df1[['contact', 'y']]
modeArray = modeExtraction['contact'].unique()
result = []

for mode in modeArray:
    modeCount = modeExtraction[modeExtraction['contact'] == mode]['y'].count()
    result.append([mode, modeCount])

resultDF = pd.DataFrame(result)
resultDF[resultDF[1] == resultDF[1].max()][0]
```

```python
# 6) HOW MANY ENTREPRENUERS IN THE LIST

print("No of Entreprenuers in the list are", df1[df1['job'] == 'entrepreneur']['job'].count())
```

```python
# 7) HOW MANY PERSONS HAVE NEGATIVE BALANCE?

print("No of people with negative balances in the list are", df1[df1['balance'] < 0]['balance'].count())
```
